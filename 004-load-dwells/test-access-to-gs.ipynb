```python
from google.oauth2 import service_account
from google_auth_oauthlib.flow import InstalledAppFlow


import dask.dataframe as dd

# Path to your downloaded OAuth client credentials
client_secrets_file = "path_to_client_secret_file"

# Scopes define the level of access requested (adjust as needed)
SCOPES = ["https://www.googleapis.com/auth/cloud-platform"]

# Launch OAuth flow to authenticate with your Google account
flow = InstalledAppFlow.from_client_secrets_file(client_secrets_file, SCOPES)
creds = flow.run_local_server(port=0)

# Use the credentials in your application
from google.cloud import storage
client = storage.Client(credentials=creds)
```

```python
import gcsfs
```

```python

def read_parquet_from_gcs(bucket_name, source_blob_name):
    """
    Reads a Parquet file directly from Google Cloud Storage using Dask.

    Parameters:
    - bucket_name: str, name of the GCS bucket.
    - source_blob_name: str, path to the Parquet file in the bucket.
    
    Returns:
    - A Dask DataFrame
    """
    # Construct the full GCS path

    fs = gcsfs.GCSFileSystem(token=client._credentials)
    
    gcs_path = f"gcs://{bucket_name}/{source_blob_name}"

    # Read the Parquet file directly from GCS
    df = dd.read_parquet(gcs_path, filesystem=fs, )
    
    return df
```

```python
# Example usage
bucket_name = "ingest-unacast-data"
folder_name = "activity_backfill/IL"
```

```python

# Read the file into a Dask DataFrame
df = read_parquet_from_gcs(bucket_name, source_blob_name)

# Display the first few rows
print(df.head())
```
