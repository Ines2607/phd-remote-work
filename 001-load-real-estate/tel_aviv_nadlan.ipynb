```python
import requests
import pandas as pd

# Base URL for the API
base_url = "https://gisn.tel-aviv.gov.il/arcgis/rest/services/IView2/MapServer/524/query"

# Parameters for the API request
params = {
    "where": "1=1",
    "outFields": "*",
    "f": "json",
    "resultRecordCount": 2000
}

# Initialize variables
# all_data = []
result_offset = 22000
more_data = True

while more_data:
    # Update the resultOffset parameter for pagination
    params["resultOffset"] = result_offset
    print(result_offset)
    # Make the API request
    response = requests.get(base_url, params=params)
    data = response.json()
    
    # Check if there are features in the response
    if "features" in data and len(data["features"]) > 0:
        all_data.extend(data["features"])
        result_offset += len(data["features"])
    else:
        more_data = False
```

```python
from shapely.geometry import Polygon
```

```python
from shapely.geometry import shape
from shapely.errors import TopologicalError
```

```python
# Extract attributes from features and convert to DataFrame
rows = []
for feature in all_data:
    attributes = feature["attributes"]
    res_geo = list(feature["geometry"].values())[0]
    if len(res_geo)>0:
        geometry = Polygon(res_geo[0])
        try:
            geom = shape(geometry) if geometry else None
            if geom and not geom.is_valid:
                geom = geom.buffer(0)  # Attempt to fix invalid geometries
            attributes['geometry'] = geom
        except (ValueError, TopologicalError):
            attributes['geometry'] = None  # Handle invalid geometries gracefully
    else:
       attributes['geometry'] = None 
    rows.append(attributes)

# Convert to GeoDataFrame
df = pd.DataFrame(rows)
gdf = gpd.GeoDataFrame(df, geometry='geometry', crs = "epsg:2039")

# # Remove rows with invalid or missing geometries
# gdf = gdf[gdf['geometry'].notnull() & gdf['geometry'].apply(lambda x: x.is_valid if x else False)]
```

```python
# Define property categories
residential = ['דירה', 'בית בודד', "קוטג' טורי", 'מגורים', 'דירת גן', "קוטג' חד משפחתי", "קוטג' דו משפחתי", 'דירת גג', 
               'דירת גג (פנטהאוז)', 'דירת נופש', 'חד משפחתי (וילה)', 'מיני פנטהאוז', 'דופלקס','דירה בבית קומות','קרקע למגורים']
commercial = ['משרד', 'חנות', 'מחסנים', 'מלאכה', 'תעשיה', 'מלונאות', 'עסק', 'בניני ציבור', 'דיור מוגן', 'מבנים חקלאיים', 
              'בית אבות', 'תחנת דלק','חניה']
mixed_use = ['מסחרי + משרדים', 'מסחרי + מגורים', 'משרדים + מגורים', 'קבוצת רכישה - קרקע מסחרי', 'קבוצת רכישה - קרקע מגורים']

# Function to categorize property type
def categorize_property(description):
    if description in residential:
        return 1  # Residential
    elif description in commercial:
        return 0  # Commercial
    elif description in mixed_use:
        return 2  # Mixed-use
    else:
        return 1  # Undefined

# Apply function to create new column
df_full['PropertyType'] = df_full['DEALNATUREDESCRIPTION'].apply(categorize_property)



```
